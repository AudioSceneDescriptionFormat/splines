# Arc Length

```python
import splines
```

```python
vertices = [
    (0, 0),
    (1, 0),
    (1, 1),
    (0, 1),
]
grid = [0, 4, 7, 8, 20]
```

```python
s = splines.CatmullRom(vertices, grid=grid, endconditions='closed')
```

```python
s.grid
```

```python
s = splines.ArcLengthReParameterizedCurve(s)
```

```python
s.grid
```

```python
s.evaluate(s.grid[-1])
```

```python
s.evaluate([0, 1, 2])
```
